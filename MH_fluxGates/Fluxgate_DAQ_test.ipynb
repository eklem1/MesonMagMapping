```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#test data take from the day we installed the fluxgates (May 10th)
# data = np.genfromtxt('labjack02_lbj0_test.csv', delimiter=',', skip_header=5)

#data from May 20 - June 20
# data = np.genfromtxt("labjack02_lbj0_220520-000000_220620-000000.csv", delimiter=',', skip_header=5)

#data from June 20 - June 29
data = np.genfromtxt("labjack02_lbj0_220620-000000_220629-000000.csv", delimiter=',', skip_header=5)

times = np.genfromtxt("labjack02_lbj0_220620-000000_220629-000000.csv", delimiter=',', skip_header=5, usecols=0, dtype='|S19')

times
```

```python
#starting from 0 instead of 1 as channel 0 = time
#time is a string rn, prob convert at some point - data every 10 s
channelNum = 0*3
chn1X = data[:,channelNum+1]
chn1Y = data[:,channelNum+2]
chn1Z = data[:,channelNum+3]

channelNum = 1*3
chn2X = data[:,channelNum+1]
chn2Y = data[:,channelNum+2]
chn2Z = data[:,channelNum+3]

timeIsh = np.arange(0, len(chn1X))

colors = ['blue', 'green', 'purple']

plt.plot(timeIsh ,chn1X, label="Channel 1 X", c=colors[0])
plt.plot(timeIsh ,chn1Y, label="Channel 1 Y", c=colors[1])
plt.plot(timeIsh ,chn1Z, label="Channel 1 Z", c=colors[2])
plt.legend()
plt.xlabel("data # ") 
plt.ylabel("?")
plt.show()

plt.plot(timeIsh ,chn2X, label="Channel 2 X", c=colors[0])
plt.plot(timeIsh ,chn2Y, label="Channel 2 Y", c=colors[1])
plt.plot(timeIsh ,chn2Z, label="Channel 2 Z", c=colors[2])
plt.legend()
plt.xlabel("data # ") 
plt.ylabel("?")
plt.show()
```
